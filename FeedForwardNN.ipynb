```python
import numpy as np
import pandas as pd
import os
import torch 
import torchvision as tvn
import sklearn as sk
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv(r"C:\Users\Andy Xie\Documents\Work\Research\Aspuru Guzik Group\Chemical Properties\data\dev\training_data.csv")
```

```python
del dataset['Unnamed: 0']
dataset = dataset.reset_index()
del dataset['index']
dataset
```

```python
X = dataset.iloc[:,1:12]
y = dataset.iloc[:,23]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train_norm = sk.preprocessing.scale(X_train)
y_train_norm = sk.preprocessing.scale(y_train)

X_test_norm = sk.preprocessing.scale(X_test)
y_test_norm = sk.preprocessing.scale(y_test)
```

```python
X_train_norm = torch.from_numpy(X_train_norm)
y_train_norm = torch.from_numpy(y_train_norm)
X_test_norm = torch.from_numpy(X_test_norm)
y_test_norm = torch.from_numpy(y_test_norm)
```

```python
X_train_norm = X_train_norm.float()
y_train_norm = y_train_norm.float()
X_test_norm = X_test_norm.float()
y_test_norm = y_test_norm.float()
```

```python
class Algernon(torch.nn.Module):
    def __init__(self):
        super(Algernon, self).__init__()
        self.ffw1 = torch.nn.Sequential(torch.nn.Linear(11,1024),torch.nn.LeakyReLU(0.2),torch.nn.Dropout(0.25))
        self.ffw2 = torch.nn.Sequential(torch.nn.Linear(1024,512),torch.nn.LeakyReLU(0.2),torch.nn.Dropout(0.25))
        self.ffw3 = torch.nn.Sequential(torch.nn.Linear(512,256),torch.nn.LeakyReLU(0.2),torch.nn.Dropout(0.25))
        self.ffw4 = torch.nn.Sequential(torch.nn.Linear(256,128),torch.nn.LeakyReLU(0.2),torch.nn.Dropout(0.25))
        self.out = torch.nn.Linear(128,1)
    
    def forward(self,x):
        x = self.ffw1(x)
        x = self.ffw2(x)
        x = self.ffw3(x)
        x = self.ffw4(x)
        x = self.out(x)
        return x

algernon = Algernon()
```

```python
optimizer = torch.optim.SGD(algernon.parameters(), lr = 0.01, momentum = 0.05)
mseloss = torch.nn.MSELoss()
```

```python
def train(epochs):
    algernon.train()
    for i in range(len(X_train_norm)):
        optimizer.zero_grad()
        output = algernon(X_train_norm[i]) 
        loss = mseloss(output, y_train_norm[i])
        loss.backward()
        optimizer.step()
        if (i + 1) % 10 == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(epochs, i, len(X_train_norm),
                                                                    100. * i / len(X_train_norm), loss.item()))

def test():
    algernon.eval()
    total_test_loss = 0
    total_correct = 0
    with torch.no_grad():
        for i in range(len(X_test_norm)):
            output = algernon(X_test_norm[i])
            test_loss = mseloss(output, y_test_norm[i])
            total_test_loss += test_loss
    avg_test_loss = total_test_loss / len(y_test_norm)
    print('\nTest set: Avg. loss: {:.4f}\n'.format(avg_test_loss))
```

```python
test()
for epoch in range(1, 4): 
    train(epoch)
    test()
```
